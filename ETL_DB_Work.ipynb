```python
# Dependencies
import pymongo
import json
import csv
import pandas as pd
import sys, getopt, pprint
from pymongo import MongoClient
```

```python
#CSV to JSON Conversion
file_to_open = "data/USvideos.csv"
df_videos = pd.read_csv(file_to_open)
df_videos.head()
```

```python
file_to_open = "data/USvideos_modified.csv"
df_videos_mod = pd.read_csv(file_to_open)
df_videos_mod.rename(columns = {"trend.publish.diff": "trend_publish_diff"}, inplace=True)
df_videos_mod.head()
```

```python
list(df_videos_mod)
```

```python
file_to_open = "data/US_category_id.json"
df_videos_id = pd.read_json(file_to_open)
df_videos_id = df_videos_id["etag"].to_frame()
df_videos_id.head()
```

```python
list(df_videos_id)
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define database and collection
db_youtube_videos = client.youtube_videos

db_youtube_videos.videos.drop()
db_youtube_videos.videos_mod.drop()
db_youtube_videos.videos_id.drop()

# Get our collections ready for insert
col_videos = db_youtube_videos.videos
col_videos_mod = db_youtube_videos.videos_mod
col_videos_id = db_youtube_videos.videos_id
```

```python
# Insert into PyMongo
# records = json.loads(df_videos.T.to_json()).values()
# db_youtube_videos.col_videos.insert_many(records)
col_videos.insert_many(df_videos.to_dict("records"))

# records = json.loads(df_videos_mod.T.to_json()).values()
# db_youtube_videos.col_videos_mod.insert_many(records)
col_videos_mod.insert_many(df_videos_mod.to_dict("records"))


# records = json.loads(df_videos_id.T.to_json()).values()
# db_youtube_videos.col_videos_id.insert(records)
col_videos_id.insert_many(df_videos_id.to_dict("records"))
```

```python
# Now get our data back out of Mongodb
df_new_videos = pd.DataFrame(list(db_youtube_videos.videos.find()))
df_new_videos
```
